### 4.2 Check and handle Missing Values

Here, first step towards treatement of missing values involves:
1. Identifying the missing values
2. Either drop the missing values or carry out imputation

Let's throw some light on the missing data:

**Types of Missing Data**

1. **MCAR - Missing Completely at Random:**
   - All variables and observations have an equal probability of being missing.
  
  
2. **MAR - Missing at Random:**
   - The probability of missing values is related to the value of the variable or other variables in the dataset.
  
3. **MNAR - Missing Not at Random:**
    - The missing data pattern is not random and is influenced by factors we may not comprehend.
  

```python
# Here we are checking missing values in the numerical and categorical columns and display them together
pd.set_option('display.max_rows', None)


missingValues = pd.concat([numerical_df.isna().sum(), categorical_df.isna().sum()], axis=1, keys=['Numerical', 'Categorical'])
print(missingValues)
```

In the course of this procedure, we are implementing imputation to handle missing values within the dataset. Specifically, for categorical data, the chosen approach involves replacing null values with the most frequently occurring values, calculated as the mode. Conversely, for numerical values, I opt to replace null entries with the mean, a strategy aimed at preserving the central tendency of the data while addressing missing values.

Imputation is chosen as the preferred strategy to retain and preserve information within the dataset rather than discarding instances with incomplete data. This approach allows us to fill in missing values, ensuring a more comprehensive and informative analysis while mitigating the potential loss of valuable information that might occur if incomplete records were simply removed.

```python
# Calculate the mode for each column
modes = categorical_df.mode().iloc[0]

# Replace NaN values with the mode for each column
categorical_df = categorical_df.apply(lambda col: col.fillna(modes[col.name]), axis=0)
```

```python
# Replace NaN values with the mean for numerical columns
numerical_df = numerical_df.apply(lambda col: col.fillna(col.mean()) if col.dtype == 'float64' or col.dtype == 'int64' else col, axis=0)
```

### Bivariate Analysis


Bivariate refers to a statistical analysis that involves the examination of the relationship between two variables.

#### Correlation Matrix

```python
import pandas as pd

# Assuming your DataFrame is named df
correlation_matrix = numerical_df.corr()

# Extracting correlation values for 'SalePrice'
saleprice_correlation = correlation_matrix['Sale_Price']

# Sorting variables based on correlation values (absolute values for both positive and negative correlations)
sorted_correlation = saleprice_correlation.abs().sort_values(ascending=False)

# Displaying the top correlated variables
print(sorted_correlation)
```

The correlation analysis reveals significant relationships between the sale price ("SalePrice") and various key features of residential properties. Notably, the overall quality of houses ("Overall Qual") demonstrates the strongest positive correlation, indicating that higher-quality homes tend to command higher sale prices. Additionally, factors such as the above-ground living area ("Gr Liv Area"), garage capacity ("Garage Cars" and "Garage Area"), and total basement and first-floor areas ("Total Bsmt SF" and "1st Flr SF") exhibit robust positive correlations, underlining the influence of living space and amenities on property values. The construction and renovation years ("Year Built" and "Year Remod/Add") also contribute positively, suggesting that newer structures and recent remodels correlate with increased sale prices. This analysis provides valuable insights into the features most closely tied to housing market dynamics, aiding in effective property valuation and market understanding.
